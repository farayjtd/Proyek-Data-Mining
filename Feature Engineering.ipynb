MISSING VALUE

```python
import pandas as pd
import numpy as np

path = r'D:\Jupyter\Proyek Data Mining\Dataset\Dataset_XAUUSD.csv'
df_simulasi = pd.read_csv(path).head(500).copy()

print(f"Dataset Awal (500 Baris)")
print(f"Jumlah baris: {len(df_simulasi)}")
```

```python
# Simulasi Menghilangkan Data Secara Acak
kolom_fitur = ['Open', 'High', 'Low', 'Close', 'Volume']

np.random.seed(None) 
for col in kolom_fitur:
    mask = np.random.choice([True, False], size=df_simulasi.shape[0], p=[0.1, 0.9])
    df_simulasi.loc[mask, col] = np.nan
```

```python
# Cek Missing Value Setelah Dirusak
print("Cek Missing Value (Setelah Data Dihilangkan Acak)")
print(df_simulasi.isnull().sum())
```

```python
# Melakukan Interpolasi Linear
df_simulasi[kolom_fitur] = df_simulasi[kolom_fitur].interpolate(method='linear')
```

```python
# Cek Missing Value Lagi Setelah Interpolasi
print("Cek Missing Value (Setelah Interpolasi Linear)")
print(df_simulasi.isnull().sum())
```

FEATURE ENGINEERING

```python
import pandas as pd

path = r'D:\Jupyter\Proyek Data Mining\Dataset\Dataset_XAUUSD.csv'
df = pd.read_csv(path)

print("Dataset telah berhasil dimuat")
print(f"Total baris: {len(df)}")
```

```python
# Menghapus duplikat berdasarkan Timestamp
df = df.drop_duplicates(subset=['Timestamp'], keep='first')

df = df.reset_index(drop=True)

print(f"Data duplikat telah dihapus.")
print(f"Jumlah baris sekarang: {len(df)}")
```

```python
# Menambahkan kolom Target
df['Target'] = df['Close'].shift(-1)

# Menambahkan kolom Keterangan (Naik, Turun, Tetap)
def tentukan_keterangan(row):
    if pd.isna(row['Target']):
        return None
    if row['Target'] > row['Close']:
        return 'Naik'
    elif row['Target'] < row['Close']:
        return 'Turun'
    else:
        return 'Tetap'

df['Keterangan'] = df.apply(tentukan_keterangan, axis=1)

# Menambahkan kolom rata-rata (Moving Average)
df['MA_5'] = df['Close'].rolling(window=5).mean()
df['MA_10'] = df['Close'].rolling(window=10).mean()
df['MA_15'] = df['Close'].rolling(window=15).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()

print("Hasil Feature Engineering")
cols_to_show = ['Timestamp', 'Close', 'Target', 'Keterangan', 'MA_5', 'MA_10', 'MA_15', 'MA_20']
print(df[cols_to_show].head(25))
```

```python
# Cek nilai kosong di setiap kolom
print("Pengecekan Nilai Kosong Setelah Feature Engineering")
print(df.isnull().sum())
```

```python
# Menghapus semua baris yang memiliki nilai kosong (NaN)
df.dropna(inplace=True)

# Reset index agar urutan baris kembali rapi dari 0
df = df.reset_index(drop=True)
```

```python
# Cek nilai kosong di setiap kolom
print("Pengecekan Ulang Nilai Kosong Setelah Feature Engineering")
print(df.isnull().sum())
```

```python
# Menampilkan 5 data pertama yang sudah bersih dan rapi
print("5 Data Pertama Setelah Feature Engineering & Cleaning")
print(df.head(5))
```
