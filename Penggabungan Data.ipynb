```python
import pandas as pd
import glob
import os

# KONFIGURASI 
folder_sumber = 'Dataset' 

nama_kolom = ['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume']

# PROSES PENGGABUNGAN
semua_file = glob.glob(os.path.join(folder_sumber, "*.csv"))

print(f"Ditemukan {len(semua_file)} file. Memulai penggabungan dengan nama kolom...")

daftar_df = []
for f in semua_file:
    df_sementara = pd.read_csv(f, sep=';', names=nama_kolom, index_col=False)
    daftar_df.append(df_sementara)

df_gabung = pd.concat(daftar_df, ignore_index=True)

# PERAPIAN FORMAT WAKTU & URUTAN 
df_gabung['Timestamp'] = pd.to_datetime(df_gabung['Timestamp'], format='%Y%m%d %H%M%S')

df_gabung = df_gabung.sort_values(by='Timestamp').reset_index(drop=True)

# PENGECEKAN HASIL
print("Selesai")
display(df_gabung.head())
```

```python
# Menampilkan dimensi data (Baris, Kolom)
print(f"Jumlah Baris: {df_gabung.shape[0]:,}")
print(f"Jumlah Kolom: {df_gabung.shape[1]}")

print("-" * 30)

# Menampilkan rincian kolom dan tipe datanya
df_gabung.info()
```
